```python
import random

def create_field():
#Создает игровое поле 3x3 с случайными числами от 1 до 9.
  field = [[random.randint(1, 9) for _ in range(3)] for _ in range(3)]
  return field

def print_field(field):
#Выводит игровое поле на консоль.
  for row in field:
    print('  '.join(str(x) for x in row))


def get_player_move(player_num, field):
#Получает ход от игрока, проверяя его корректность.
  while True:
    print(f"Игрок {player_num}, ваш ход.")
    try:
      row, col = map(int, input("Введите координаты хода (строка, столбец): ").split())
      row -= 1  # Индексы в списке начинаются с 0
      col -= 1
      if 0 <= row <= 2 and 0 <= col <= 2 and field[row][col] != 0:
        return row, col
      else:
        print("Некорректный ход. Попробуйте еще раз.")
    except ValueError:
      print("Некорректный ввод. Введите два числа через пробел.")

def check_game_over(field):
#Проверяет, завершена ли игра.
  for row in field:
    if any(x != 0 for x in row):
      return False
  return True

def play_maxit():
#Запускает игру.
  field = create_field()
  player1_score = 0
  player2_score = 0
  current_player = 1

  print("Добро пожаловать в игру Максит!")
  print_field(field)

  while not check_game_over(field):
    row, col = get_player_move(current_player, field)

    if current_player == 1:
      player1_score += field[row][col]
    else:
      player2_score += field[row][col]

    field[row][col] = 0  # Вычеркиваем число

    print_field(field)

    if current_player == 1:
      current_player = 2
    else:
      current_player = 1

  print("\nИгра окончена!")
  print(f"Счет: Игрок 1 - {player1_score}, Игрок 2 - {player2_score}")

  if player1_score > player2_score:
    print("Победил Игрок 1!")
  elif player1_score < player2_score:
    print("Победил Игрок 2!")
  else:
    print("Ничья!")

if __name__ == "__main__":
  play_maxit()



'''
1) create_field(): Создает игровое поле с случайными числами от 1 до 9.

2) print_field(field): Выводит поле на экран.

3) get_player_move(player_num, field):
1. Запрашивает ход у игрока с помощью input().
2. Проверяет корректность ввода, убеждаясь, что координаты находятся в пределах поля и что клетка не пустая.
3. Возвращает координаты хода.

4) check_game_over(field): Проверяет, остались ли на поле невычеркнутые числа.

5) play_maxit():
1. Проверяет игровое поле, очки игроков и текущего игрока.
2. Запускает цикл, пока игра не закончится.
3. Внутри цикла:
a. Вызывает get_player_move(), чтобы получить ход.
b. Обновляет очки игрока.
c. Вычеркивает выбранное число на поле.
d. Переключает ход к другому игроку.

6) После завершения игры выводит результаты.
'''
```
